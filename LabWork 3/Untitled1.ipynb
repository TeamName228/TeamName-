```python
import requests
from bs4 import BeautifulSoup
```

```python
def parse_studfile(url):
    response = requests.get(url)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')
    
    meta_descriptions = [
            meta['content'].strip()
            for meta in soup.find_all('meta', {'property': 'og:description'})
            if 'content' in meta.attrs
        ]

    paragraphs = [
        p.text.strip().replace('\n', ' ')
        for p in soup.find_all('p', {'align': 'center'})
    ]
    
    content = [
        p.text.strip()
        for p in soup.find_all('p')  # Находим все теги <p>
        if not p.attrs and p.text.strip() # Проверяем, что у тега нет атрибутов и он не пустой
    ]


    return {
        'meta_description': meta_descriptions,
        'paragraphs': paragraphs,
        'content': content
    }
```

```python
urls = [
    'https://studfile.net/preview/4380990/',
    'https://studfile.net/preview/7331425/',
    'https://studfile.net/preview/9518226/',
    'https://studfile.net/preview/1487560/',
    'https://studfile.net/preview/16532039/',
    'https://studfile.net/preview/7196174/',
    
    
]
```

```python
for url in urls:
    print(" ")
    print(url)
    result = parse_studfile(url)
    print('\nМетаданные: ')
    for desc in result['meta_description']:
        print(desc) 
        
    print('\nПараграф: ')
    for p in result['paragraphs']:
        print(p) 
        
    print('\nКонтент: ')
    for c in result['content']:
        print(c)
        
```
