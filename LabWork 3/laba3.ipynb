```python
dat2 = []
dat2.append(data_info1[27])
print(dat2)
```

```python
data_info_words = dat2[0].split()
print(data_info_words)
```

```python
filtered_words_without_digit = [word for word in data_info_words if not word.isdigit()]
print(filtered_words_without_digit)
```

```python
words_to_remove = ['Заключение', 'Список', 'литературы', 'Задание:', 'Разработать', 'корпоративную...']
filtered_words_without_w = [word for word in filtered_words_without_digit if word not in words_to_remove]
print(filtered_words_without_w)
```

```python
# Функция для проверки, начинается ли строка с числа (например, "1.", "1.1")
def starts_with_number(s):
    return any(char.isdigit() for char in s.split('.')[0])

# Новый список для объединенных предложений
combined_sentences = []
current_sentence = []

# Проходим по списку слов
for word in filtered_words_without_w:
    if starts_with_number(word):
        # Если текущее предложение не пустое, добавляем его в результат
        if current_sentence:
            combined_sentences.append(' '.join(current_sentence))
            current_sentence = []
        # Начинаем новое предложение с текущего слова
        current_sentence.append(word)
    else:
        # Продолжаем добавлять слова в текущее предложение
        current_sentence.append(word)

# Добавляем последнее предложение, если оно есть
if current_sentence:
    combined_sentences.append(' '.join(current_sentence))

# Выводим результат
filtered_data_info = combined_sentences
print(filtered_data_info)
```
