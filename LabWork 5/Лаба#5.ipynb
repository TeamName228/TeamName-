```python
import json
import ollama

def load_dataset(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        return json.load(f)

def generate_explanatory_note(model, topic):
    prompt = f"Сгенерируй пояснительную записку по теме: {topic}"
    response = ollama.chat(model=model, messages=[{"role": "user", "content": prompt}])
    return response['message']['content']

def main():
    dataset_path = "datasets/dataset_deepseek.json" 
    model_name = "deepseek-r1:14b"
    
    dataset = load_dataset(dataset_path)
    
    for item in dataset[:3]:
        topic = item['input']
        print(f"\nТема: {topic}")
        print("Генерируем пояснительную записку...\n")
        result = generate_explanatory_note(model_name, topic)
        print(result)
    
if __name__ == "__main__":
    main()
```
